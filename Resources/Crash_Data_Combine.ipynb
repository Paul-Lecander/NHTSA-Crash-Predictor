```python
# Dependencies
import pandas as pd
```

# ACCIDENT

```python
# read CSV accident 2020
accident_20_df=pd.read_csv('CRSS2020CSV/accident.csv')
accident_20_df.head()
```

```python
# SELECT RETAINED COLUMNS ACCIDENT 2020
accident_20_brief_df= accident_20_df[['CASENUM','URBANICITY','VE_TOTAL','VE_FORMS','PERMVIT','NUM_INJ','MONTH',
                                      'YEAR','DAY_WEEK','HOUR','ALCOHOL','MAX_SEV','WRK_ZONE','LGT_COND','WEATHER']]
accident_20_brief_df.head()
```

```python
# read CSV accident 2019
accident_19_df=pd.read_csv('CRSS2019CSV/accident.csv')
accident_19_df.head()
```

```python
# SELECT RETAINED COLUMNS ACCIDENT 2019
accident_19_brief_df= accident_19_df[['CASENUM','URBANICITY','VE_TOTAL','VE_FORMS','PERMVIT','NUM_INJ','MONTH',
                                      'YEAR','DAY_WEEK','HOUR','ALCOHOL','MAX_SEV','WRK_ZONE','LGT_COND','WEATHER']]
accident_19_brief_df.head()
```

```python
accident_20_brief_df.info()
```

```python
accident_19_brief_df.info()
```

```python
# APPEND DataFrames
accident_df = accident_19_brief_df.append(accident_20_brief_df)
accident_df.head()
```

```python
accident_df.info()
```

```python
# Export CSV
accident_df.to_csv('accident_19_20.csv')
```
